```python
import numpy as np
import seaborn as sns

def where_dropped(level: int, number: int) -> dict:
    data = {}
    while (number > 0):
        temp_level = level
        sum = 0
        while (temp_level > 0):
            left_or_right = np.random.choice([-1,1])
            sum += left_or_right
            temp_level = temp_level - 1
        if sum in data:
            data[sum] += 1
        else:
            data[sum] = 1
        number = number - 1
    return data

data = where_dropped(11, 10000)
print(data)
sns.barplot(x=list(data.keys()), y=list(data.values()));
```
